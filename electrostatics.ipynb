- Legge di Coulomb
$$F = \frac{1}{4\pi \epsilon_0} \frac{Q_{tot}}{r^2}$$

- forze esercitate da ciascuna carica:
$$F_1 = F_2 = \frac{1}{4 \pi \epsilon_0} \frac{2 q^2}{a^2 + h^2}$$
- modulo della forza risultante
$$F = 2 F_1 \cos(\theta)$$
- posto che:
$$\cos(\theta) = \frac{h}{a^2 + h^2}$$

per cui:
$$F = \frac{1}{4 \pi \epsilon_0}{4 q^2}{a^2 + h^2} \frac{h}{(a^2 + h^2)^{1/2}}= \frac{q^2}{\pi \epsilon_0} \frac{h}{(a^2 + h^2)^{1/2}}$$

- $F = F(h)$, per cui, luogo dei punti dove F è maggiore:
$$\frac{dF(h)}{dh} = \frac{q^2}{\pi \epsilon_0} \frac{(a^2 + h^2)^{3/2} - 3h^2(a^2 + h^2)^{1/2}}{(a^2 + h^2)^3} = \frac{q^2}{\pi\epsilon_0}\frac{a^2 - 2h^2}{(a^2 + h^2)^{5/2}} = 0$$
- per cui:
$$\frac{dF(h)}{dh} = 0$$
per 
$$a^2 - 2h^2 = 0$$
per
$$h^2 = a^2/2$$

- posto che:
$$h = \sqrt{x^2 + z^2}$$
allora
$$x^2 + z^2 = a^2/s$$
è il luogo dei punti dove la F è massima

**Numericamente**

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


# dati
eps_0 = 8.85*10**(-2)                     # [C2/Nm2] costante dielettrica nel vuoto
q = 10**(-2)                             # [C]   carica elettrica
a = 1                                     # [cm] distanza cariche dall'origine del riferimento
z = 2                                     # [cm] distanza in direzione z
x = 2                                     # [cm] distanza in direzione x

h = 10                                    # [cm] distanza in verticale
# determinazione della forza
F1 = 1/(4*np.pi * eps_0) * (2*q**2)/(a**2 + h**2)
# risultante della forza 
theta = np.arccos(h/((a**2 + h**2)**(1/2)))
F = 2*F1*np.cos(theta)
print('La forza esercitata sulla carica è pari a (N): ', np.around(F, decimals=10))
```

- In questo tipo di problemi torna utile l'**analisi infinitesimale**

- contributo della carica infinitesima dq sul tratto dl:
$$dE = \frac{1}{4  \pi \epsilon_0} \frac{dq}{r^2} = \frac{1}{4 \pi \epsilon_0} \frac{\lambda dl}{r^2}$$
- posto che:
$$dq = \lambda dl$$
- è la carica infinitesima

- **componenti**
$$dE_p = dE\sin{\theta}$$
$$dE_n = dE\cos{\theta}$$
- posto che
$$r = d = \cos{\theta}$$
- e che
$$l = d \tan{\theta}$$
- allore
$$dl = \frac{d}{\cos{\theta}^2} d\theta$$

- per cui
$$dE_p = \frac{\lambda}{4 \pi \epsilon_0 d} \sin{\theta} d\theta$$
- e
$$dE_n = \frac{\lambda}{4 \pi \epsilon_0 d} \cos{\theta} d\theta$$

- integrando:
$$E_p = \int_{-\theta_1}^{\theta_2} \frac{\lambda}{4 \pi \epsilon_0 d} \sin{\theta} d \theta = \frac{\lambda}{4 \pi \epsilon_0 d} [ - \cos{\theta} ]_{-\theta_1}^{\theta2} = \frac{\lambda}{4 \pi \epsilon_0 d} (\cos{\theta_1 } - \cos{\theta_2})$$

$$E_n = \int_{-\theta_1}^{\theta_2} \frac{\lambda}{4 \pi \epsilon_0 d} \cos{\theta} d \theta = \frac{\lambda}{4 \pi \epsilon_0 d} [ - \sin{\theta} ]_{-\theta_1}^{\theta2} = \frac{\lambda}{4 \pi \epsilon_0 d} (\sin{\theta_1 } + \sin{\theta_2})$$

modulo del campo elettrostatico:
$$E = \left( E_p^2 + E_n^2\right)^{1/2}$$

ovvero:
$$E = \frac{\lambda}{4 \pi \epsilon_0 d}[(\cos{\theta_1} - \cos{\theta_2})^2 + (\sin{\theta_1} + \sin{\theta_2})^2]^{1/2} = \frac{\lambda}{4 \pi \epsilon_0 d}[2 - 2 \cos{(\theta_1 + \theta_2)}]^{1/2}$$

```python
# Numericamente
# dati
dl = 0.01                                 # [m]  lunghezza di un tratto infinitesimo di filo
dq = 10**(-6)                             # [C]  carica infinitesima del tratto inifinitesimo di filo
d = 0.1                                   # [m]  distanza perpendicolare del punto P dal filo
eps_0 = 8.85*10**(-2)                     # [C2/Nm2] costante dielettrica nel vuoto
lam = dq/dl                              # [C/m] densità di carica lineare  
l = 1                                     # [m]  lunghezza complessiva del filo
AH = 0.2                                  # [m]  distanza dall'estremo sinistro alla perpendicolare in H
HB = l - AH                               # [m]  distanza dall'estremo destro alla perpendicolare in H

r1 = np.sqrt(d**2 + AH**2)                # [m]  r1 distanza dall'estremo di sinistra al punto P
r2 = np.sqrt(d**2 + HB**2)                # [m]  r2 distanza dall'estremo di destra al punto P
teta1 = np.arccos(d/r1)
teta2 = np.arccos(d/r2)

# Determino il valore della soluzione ottenuto per via analitica
E = lam/(4*np.pi*eps_0 *d )*(2 - 2*np.cos(teta1 + teta2))**(1/2)
print('Il valore calcolato per via analitica è[N/C]: ', np.around(E, decimals=6))
```

```python
# Integrando rispetto a teta1 e teta2
from scipy.integrate import quad
def dEp(teta):
    return lam/(4*np.pi*eps_0*d)*np.sin(teta)

I1 = quad(dEp, -teta1, teta2)

def dEn(teta):
    return lam/(4*np.pi*eps_0*d)*np.cos(teta)

I2 = quad(dEn, -teta1, teta2)

Ep = I1[0]
En = I2[0]
E = np.sqrt(Ep**2 + En**2)
print('Il valore del campo calcolare integrando rispetto a teta è[N/C]: ', np.around(E, decimals=6))
```

```python
E = []
I  = np.arange(0, 10, 0.1)
for l in I:
    AH = 0.2*l                                  # [m]  distanza dall'estremo sinistro alla perpendicolare in H
    HB = l - AH 
    r1 = np.sqrt(d**2 + AH**2)                # [m]  r1 distanza dall'estremo di sinistra al punto P
    r2 = np.sqrt(d**2 + HB**2)                # [m]  r2 distanza dall'estremo di destra al punto P
    teta1 = np.arccos(d/r1)
    teta2 = np.arccos(d/r2)
    def dEp(teta):
        return lam/(4*np.pi*eps_0*d)*np.sin(teta)
    I1 = quad(dEp, -teta1, teta2)
    Ep = I1[0]
    def dEn(teta):
        return lam/(4*np.pi*eps_0*d)*np.cos(teta)
    I2 = quad(dEn, -teta1, teta2)
    En = I2[0]
    result = np.sqrt(Ep**2 + En**2)
    E.append(result)
```

```python
# Plotting
plt.plot(I,E)
plt.xlabel('lunghezza del filo [m]')
plt.ylabel('Valore del campo elettrostatico [N/C]')
plt.title('Campo elettrostatico generato da un filo in relazione alla sua lunghezza')
plt.grid()
plt.show()
```

# Energia, condensatori, dielettrici

- determinazione dell'energia elettrostatica di una carica distribuita uniformemente su una superficie sferica

**densità di carica superficiale**:
$$\sigma = \frac{dq}{dS}$$
- per una sfera $S = 4 \pi R^2$
- per cui:
$$Q = 4 \pi R^2 \sigma$$

tale carica in realtà si può pensare distribuita in uno spessore infinitesimo **dr**
- campo all'interno: **$E_i = 0$**
- campo al'esterno: **Teorema di Coulomb** $\rightarrow$ $E_e = \frac{\sigma}{\epsilon_0}$
- punti interni allo spessore **dr**: $E^* = (E_i + E_e)/2 = \frac{\sigma}{2 \epsilon_0}$

sotto l'azione di tale campo la carica infinitesima dq è soggetta alla forza:
$$F = E^* dq = \frac{1}{2} \frac{\sigma}{\epsilon_0} \sigma dS = \frac{1}{2} \frac{\sigma^2}{\epsilon_0} dS$$
- da cui si può definire una pressione elettrostatica:
$$P_e = \frac{dF}{dS} = \frac{1}{2} \frac{\sigma^2}{\epsilon_0}$$

- **l'effetto della pressione elettrostatica ** è quello di favorire la tendenza ad allontanarsi delle cariche, ovvero di aumentare la superficie su cui le cariche si distribuiscono.

calcolo del lavoro necessario per variare di dr il raggio, o dV il volume:
$$dW = P_e dV = \frac{1}{2} \frac{\sigma^2}{\epsilon_0} dV = \frac{1}{2} \epsilon_0 \frac{\sigma}{\epsilon_0}\frac{\sigma}{\epsilon_0} dV = \frac{1}{2} \epsilon_0 E^2 dV$$

- pensando di portare il valore del raggio all'infinito:
$$W = \int_{V_1}^{V_2} \frac{1}{2} \frac{\sigma^2}{\epsilon_0} dV = \int_{R}^{\infty} \frac{1}{2 \epsilon_0} \frac{Q^2}{(4 \pi r^2)^2} 4 \pi r^2 dr = \frac{1}{2} \frac{Q^2}{4 \pi \epsilon_0 R}$$

altra possibile interpretazione di questo lavoro:
$\textit{il lavoro che è stato necessario compiere contro le forze del campo per portare le cariche Q dall'infinito sulla sfera di raggio R, e quindi come l'energia elettrostatica associata alla distribuzione sferica di carica.

il termine $\frac{dW}{dV} = \frac{1}{2} \epsilon_0 E^2$, rappresenta la **densità di energia**.

posto $C = 4 \pi \epsilon_0 R$ $\implies$ $W = \frac{Q^2}{2C}$ $\implies$ la costante C rappresenta la **capacità della sfera**

- **la carica si distrivuisce quindi sulla superficie di una sfera conduttrice, sia che essa sia cava, sia che essa sia piena**

```python
# Numericamente
# Dati
eps_0 = 8.85*10**(-2)                           # [C2/Nm2]  costante dielettrica nel vuoto
sigma = 5 * 10**(-1)                           # [C/m2]     densità di carica superficiale

# determinazione della pressione elettrostatica esercitata sulla superficie
Pe = 1/2. * sigma**2/eps_0
print('La pressione elettrostatica esercitata sulla superficie è pari a [N/m2]: ', np.around(Pe,decimals=4))
```

```python
# il lavoro complessivamente necessario per variare il volume dal raggio R all'infinito:
R = 0.01                                        # [cm] raggio della sfera
Q = 4 * np.pi*R**2 * sigma                      # [C]  carica complessiva della sfera
W = 1/2. * (Q**2)/(4*np.pi*eps_0 * R)           # [J/m2] lavoro per unità di superficie compiuto per portare le cariche dall'infinito sulla sfera
print('Il lavoro compiuto per portare le cariche dall\'infito alla sfera è pari a[J]: ', np.around(W,decimals=5))
```

```python
# determinazione della capacità della sfera
C = 4*np.pi*eps_0*R                             # [C2/J]   capacità della sfera
print('La capacità della sfera è pari a[C2/J]: ', np.around(C, decimals=5))
```

Il ciclotrone è un tipo di acceleratore di particelle; ci sono dei magneti che fanno compiere una traiettoria circolare alla particella, la quale viene accelerata da un campo elettrico variabile. Il campo magnetico è costante cosicchè la particella accelerando aumenta il suo raggio di curvatura, secondo la legge:
$$R = \frac{m v}{q B}$$

e dopo alcuni giri viene sbalzata fuori. Il campo elettrico variabile viene generato da una d.d.p. . 

Dal punto di vista pratico le applicazioni del ciclotrone sono quelle di acceleratore delle particelle che bombardano i tessuti neoplastici. Dato che:
$$R = \frac{m v}{q B}$$
allora
$$B = \frac{m v}{q R}$$

- $\overline{M}$ : è il momento risultante delle forze agenti sul sistema
- $\oint$ : è l'integrale di linea eseguito su tutto il circuito $\rightarrow$ **circuitazione**

anche il filo produce un campo magnetico, con quale legge?

**Legge di Biot-Savart**: legge fondamentale della magnetostatica

prima legge di Laplace per circuiti filiformi:
$$dB = \frac{\mu_0}{4 \pi} i \frac{dl \times \vec{r}}{r^3} = \frac{\mu_0}{ 4 \pi} i \frac{ dl \times \vec{r}}{r^2}$$

ovvero:
- **legge di Biot-Savart**
$$\vec{B} = \frac{\mu_0}{4 \pi} \oint_l i \frac{dl \times \vec{r}}{r^3}$$

per circuiti non filiformi
$$i = \int_S \vec{J} d \vec{S}$$

allora:
- **legge di Biot-Savart**
$$\vec{B} = \frac{\mu_0}{4 \pi} \int_{\tau} \frac{\vec{J} \times \vec{r}}{r^3} d \tau$$

- $\mu_0$ è la permeabilità magnetica nel vuoto
- $\mu_0 = 12.56 10^{-7}$ $[\frac{wb}{m^2} \frac{m}{A}]$ $ = 4 \pi 10^{-7}$ $[\frac{Volt}{m} \frac{s}{A}]$ $ = 4 \pi 10^{-7} \frac{\Omega s}{m}$ $ = 4 \pi \frac{N}{A^2}$

```python
Image('C:\\Users\\nicol\\Desktop\\FISICA2\\IMMAGINI\\induzione9.PNG', width=400)
```

da Biot-Savart:
$$B2 \pi = \mu_0 i$$    

essendo
$$B 2 \pi d = \int B d \vec{l}$$

allora essa è la circuitazione di $\vec{B}$

**teorema della circuitazione di Ampère**
$$\oint \vec{B} d \vec{l} = \mu_0 i$$

La circuitazione del campo magnetico lungo una qualunque linea chiusa, è uguale a $\mu_0$ per la somma delle correnti concatenate:
$$\oint \vec{B} d \vec{l} = \mu_0 \Sigma_{i, concatenate}$$

per il teorema di Stokes:
$$\oint_l \vec{B} d \vec{l} = \int_S \mu_0 \vec{J} dS$$

allora:
$$\oint \vec{B} \cdot d \vec{l} = \int_S \vec{\nabla} \times \vec{B} d \vec{S}$$


e dato che:
$$i = \int_S \vec{J} \cdot d \vec{S} = \int_S \vec{J} \cdot \hat{n} dS$$

$$\implies$$

$$\int_S \vec{\nabla} \times \vec{B} d \vec{S} = \int_S \mu_0 \vec{J} d \vec{S}$$

ovvero:
$$\vec{\nabla} \times \vec{B} = \mu_0 \vec{J}$$

che rappresenta la **quarta equazione di Maxwell**

$\implies$ $\vec{B}$ **NON È IRROTAZIONALE!!**

essendo la seconda equazione di Maxwell:
$$\vec{\nabla} \cdot \vec{B} = 0$$

$\implies$ il campo $\vec{B}$ è un **campo solenoidale** ma **NON È CONSERVATIVO!!**

Non essendo conservativo noi non ci chiediamo di trovare il potenziale magnetico; perchè il potenziale magnetico **NON ESISTE!**
